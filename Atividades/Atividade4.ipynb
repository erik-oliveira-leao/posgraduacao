```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

# model & tokenizer
model_name = "pierreguillou/bert-large-cased-squad-v1.1-portuguese"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
print(text)
```

```python
questionAndAnswer = pipeline("question-answering", model=model, tokenizer=tokenizer)

pergunta1 = "Qual é o maior problema apontado para o carro?"
pergunta2 = "Há um otimismo para uma possível melhora das condições do carro?"
pergunta3 = "As ondulações da pista afetam de alguma maneira o chassis do carro?"

resposta1 = questionAndAnswer(question=pergunta1, context=text)
resposta2 = questionAndAnswer(question=pergunta2, context=text)
resposta3 = questionAndAnswer(question=pergunta3, context=text)

print(pergunta1,"\n\tResposta:", resposta1['answer'])
print(pergunta2,"\n\tResposta:", resposta2['answer'])
print(pergunta3,"\n\tResposta:", resposta3['answer'])
```
